##Tratamiento de Datos Faltantes (Tratamiento Condicional)

- **~** - permite seleccionar lo opuesto

```python
datos_residencial
```

```python
# inmuebles residenciales Mantenimiento.NaN
datos_residencial[datos_residencial['Mantenimiento'].isnull()].shape[0]
```

```python
# Series booleana con inmuebles residenciales, Tipo.Departamento & Mantenimiento.NaN
datos_departamentos = (datos_residencial['Mantenimiento'].isnull()) & (datos_residencial['Tipo'] == 'Departamento')
datos_residencial[datos_departamentos]
```

```python
# eliminando observaciones de inmuebles con Mantenimiento.NaN & son Tipo.Departamento (diferentes de) ~ Tipo.Departamento & Mantenimiento.NaN
antes_datos_residencial = datos_residencial.shape[0] # observaciones antes de eliminar
datos_residencial = datos_residencial[~datos_departamentos] # se actualiza en datos_residencial lo opuesto(~) datos_departamentos
despues_datos_residencial = datos_residencial.shape[0] # observaciones despues de eliminar
antes_datos_residencial - despues_datos_residencial # inmuebles residenciales sin Departamentos.NaN
```

```python
# contabilizamos los registros Mantenimiento.NaN restantes
datos_residencial[datos_residencial['Mantenimiento'].isnull()].shape[0]
```

```python
# reemplazamos los valores NaN en Mantenimiento e Impuesto
datos_residencial = datos_residencial.fillna({'Mantenimiento': 0, 'Impuesto': 0}) # diccionario con valores a sustituir = 0
```

```python
# contabilizamos registros Mantenimiento.NaN, ahora son 0
datos_residencial[datos_residencial['Mantenimiento'].isnull()].shape[0]
```

#Métodos de Interpolación

Métodos para limpieza de valores NaN

- **.fillna()** - reemplaza los valores NaN por cualquier valor (), siempre y cuando inplace = True

- **.fillna(method = 'ffill')** - reemplaza el valor NaN con el valor anterior
- **.fillna(method = 'bfill')** - reemplaza el valor NaN con el valor siguiente

- **.fillna(*df*.mean())** - reemplaza los valores NaN con el valor de la media - **MÉTODO MÁS UTIIZADO PARA SUSTITUIR NaN**

- **.fillna(method = 'ffill', *limit = 1*)** - limit = x, limita el número de veces que se sustituirán valores NaN

```python
data_inter = [0.5, None, None, 0.52, 0.54, None, None, 0.59, 0.6, None, 0.7]
s_inter = pd.Series(data_inter)
s_inter
```

```python
# fillna
s_inter.fillna(0)
```

```python
# ffill
s_inter.fillna(method = 'ffill')
```

## *Boxplot: diagrama de caja*

Utiliza algunas medidas estadísticas para **cortar nuestros datos en algunos cuartiles** e identificar cuáles puntos o cuáles datos o cuáles líneas de nuestro banco están fuera de estos cortes.

Crea una línea **mediana** que divide los datos a la mitad, 50% de los datos los coloca para un lado y el otro 50% los coloca del otro lado.

También crea **cuartiles**. El **Q1** selecciona el primer **25%** de nuestros datos, la **mediana** el **50%**, y el **Q3** el **75%** restante.

Encima de estas medidas va a crear un límite inferior y un límite superior, donde todas las líneas o los puntos que están fuera son nuestros **outliers**.

El **intervalo intercuartil(IIQ),** que es la resta del Q3 menos el Q1


boxplot.png

```python
# tiene que ser una columna que se pueda medir
datos_residencial.boxplot(['Valor'], figsize = (5,7)) # la mayoria de los datos está entre los 500,000 USD
```

```python
valor
```

```python
# para eliminar los outliers se deben instanciar los quartiles
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3-Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
# filtro outliers
seleccion_outliers = ((valor >= limite_inferior) & (valor <= limite_superior))
```

```python
area2 = plt.figure()
# (nº de líneas en el área, nº de columnas en el área, posición del gráfico)
g5 = area2.add_subplot(1, 2, 1)
g6 = area2.add_subplot(1, 2, 2)

grupo1 = datos_sin_outliers.groupby('Tipo Agrupado')['Valor']
label = grupo1.count().index
valores = grupo1.count().values
g5.pie(valores, labels = label, autopct='%1.1f%%')
g5.set_title('Total de Inmuebles por Tipo Agrupado')

grupo2 = datos_sin_outliers.groupby('Tipo')['Valor']
label = grupo2.count().index
valores = grupo2.count().values
g6.pie(valores, labels = label, autopct='%1.1f%%', explode = (.1, .1, .1, .1, .1, .1)) # explode = ( .1*6, 6 tipos de inmuebles)
g6.set_title('Total de Inmuebles por Tipo')
```
